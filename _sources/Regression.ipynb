# Linear Model (LM)

Важный класс моделей машинного обучения – линейных. **Линейные модели** –– предполагааюь, что опредеяемы критерий линейно зависит от признаков описывающих ообъект или процесс. На данный момент развития машинного обучения и искусственного интелекта в целом можно строить довольно сложные модели, нейросети, но практически везде в основе лежат линейные модели. Они имеют большое кол-во преимуществ, например, такие как относительная простота вычислений и интепретация результатов, эффективность для описания многих процессов. Линейные модели используются в прогнозе цен, построении вероятностных моделей в кредитном скоринге, трейдинге, и т.д.

Линейные модели описываются довольно простым линейным уравнением, где $y$ - целевая переменная, $w$ - вектор весов модели, $X$ - матрица наблюдений,  а $e$ - ошибка.$$y_i = \sum\limits^{m}_{j=1}w_j X_{ij} + e_i$$

<p style="align: center;"><img align=center src="https://wiki.loginom.ru/images/linear-regression.svg"  width=900></p>

**Плюсы:**

    1. Скорость и простота
    2. Интерпретируемость модели. Линейная модель являеся прозрачной и понятной для аналитика. По полученным коэффициентам регресси можно судить о том, как тот или иной факутор влияет на результат, сдалеть на это основе дополонительные поезные выводы.
    3. Широкая применимость. Большое количество реальных процессов в экономике и бизнесе можно с достаточной точностью описать линейными моделями.
    4. Изученность данного подхода. Для линейной регрессии известны типичные проблемы (например мультиколлинерность) и их решения, разработна и реализованы тесты оценки стиатической значимости получаемых моделей.
    
**Минусы:**
    
    1. Не могут отследить сложные зависимости в данных.
    2. Проблемы мультиколлинеарности.

#### Метод максимального правдоподобия
<p id="2">Если процесс описывается с помощью линеной модели, то необходимо определить параметры $\alpha$ модели, для максимально эффективной точности. На языке математики это означает, что нужно задать вероятность $$p(y|x,\alpha)$$ того, что модель с параметрами $\alpha$ и входными данными $x$ будет максимальной.</p>
Введем функцию правдоподобия, от которой нужно потребовать, чтобы произведение вероятностей по всем возможным $x$ с параметрами $\alpha$ было максимально.
$$W(\alpha) = \prod\limits_{i}p(x_i, \alpha)$$
Тогда прологарифмировав выражение, получаем функцию масимального правдоподобия. $$L(\alpha) = \sum\limits_i \log p(x_i,\alpha)$$
Чтобы максимизировать $L(\alpha)$ необходимо минимизировать среднеквадратичную ошибку между прогнозируемыми и фактическими значениями.

# Обучение алгоритма/модели машинного обучения

Для обучения модели определяются:

    1. Функция потерь.
    2. Технические метрики,с помощью которых определяют качесвто обучения модели.
    3. Бизнес метрики, с помощью которых поредеют бизнес эффект модели.
    
    
**Функция потерь** ––  это мера количества ошибок, которые наша модель делает на наборе данных.

Функция потерь задает гипрповерхность в прострвен параметров модели (пространство функций потерь) Минимзурюя функцию потерь мы получаем модель, которая дает минимальное отклонение предсказанных данных от реальных. 

Ошибки, которые совершает модель:

1. Ошибки **обучения** (на тренироочных данных)
2. Ошибки **обобщения** (на проверочных данных)

# Линейная регрессия

Регрессия –– уравнение связи искомого критерия с признаками.

Линейна регрессия – связь между искомым критерием и параметрами линейная

$$
y_i = \sum\limits_{j=1}^{m} w_j X_{ij} + e_i
$$

$Y$ –– целевая переменная
$W$ –– вектор весов модели
$X$ –– матрица наблюдений
$e$ –– ошибка модели

```python
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np

plt.style.use('_mpl-gallery')

# Make data
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
# R = np.sqrt(X + Y)
Z = (-X**2 - Y**2)**2

# Plot the surface
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, Z, vmin=Z.min() * 2, cmap=cm.Blues)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```

# $$ \log p ( \vec y | X, \vec w  ) = -\frac{n}{2} \log 2 \pi \sigma^2 - \frac{1}{2 \sigma^2} \sum\limits_{i=1}^{n} \big( y_i - \vec w ^{T} \vec x_i \big)^2 $$

# Функция потерь

# $$ \mathbf L (X, \vec y, \vec w ) = \frac{1}{2\sigma^2} \big( y_i - \vec w^{T} \vec x_i \big)^2  .$$

В каких случаях мы можем использовать линейную регресию? Какие данные мы можем использовать?

**Требования к данным.**
1. Линейная зависимость целевой переменной
2. Нормальное распределение остатков
3. Постоянная изменчивость остатков. Гомоскедастиность.
4. независимость признаков $ corr(x_i, x_j)=0 $

## Виды регрессии

### 1. Полиномиальная 
#    $$ f(X) = \sum\limits_{i=1}^{n} \sum\limits_{j=1}^{m} w_i x_i^j + b $$

### 2. Логорифмическая 
#    $$ f(X) = \sum\limits_{i=1}^{n}  w_i \ln( x_i) + b $$


### 3. Экспоненциальная 
#    $$ f(X) = \sum\limits_{i=1}^{n} e^ {w_i x_i} + b $$

### 4.Гиперболическая
#    $$ f(X) = \sum\limits_{i=1}^{n} \cdot \frac{1}{x} + b $$

### Полиномиальная регрессия
<p id="4">Не всегда данные могут быть описаны (достаточно эффективно) линейной моделью. В некоторых задачах часто строятся полиномиальные модели. Здесь появляется дополнительный гиперпараметр - максимальная степень, в  которой могут находится параметры для описанния данных. Хороший пример как работает полиномиальная и линейная регрессия на нелинейных данных

```python
from sklearn.metrics import mean_squared_error, r2_score # подгружаем метрику mse и r2
from sklearn.preprocessing import PolynomialFeatures
```

```python
model = LinearRegression() 
model.fit(X, Y)
y_pred = model.predict(X)
plt.figure(figsize=(12,9))

# считаем метрику MSE и R2
rmse = np.sqrt(mean_squared_error(Y, y_pred))
r2 = r2_score(Y, y_pred)
print('RMSE:', rmse)
print('R2:', r2)
```
